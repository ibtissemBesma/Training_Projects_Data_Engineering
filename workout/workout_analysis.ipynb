```python
def extract_data(file_path):
    return pd.read_csv(file_path)
```

```python
workout=extract_data('data/workout.csv') ##  month	| workout_worldwide
```

```python
def get_year(period):
    return period.split('-')[0]
year=workout.loc[workout['workout_worldwide']==workout['workout_worldwide'].max(),'month'].apply(get_year)
year=int(year.iloc[0])
year
```
